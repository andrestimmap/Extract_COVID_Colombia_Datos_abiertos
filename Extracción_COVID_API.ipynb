### Estandarización atención

```python
# en columna recuperado, se estandariza

df['recuperado'].replace(
    to_replace=['fallecido'],
    value='Fallecido',
    inplace=True
)

#Se ve la sumatoria de los estados, según columna recuperado

Estado = df.groupby(['recuperado']).sum().reset_index()
Estado
```

```python
#Se conviente edad string a numerico

df['Edad'] = pd.to_numeric(df["edad"])
```

```python
df['Mayores_Hombres'] = 0
df['Mayores_Mujeres'] = 0
```

```python
def Mayores():
    for i in range(contarFilas):
        if df['sexo'][i] == 'Masculino' and df['Edad'][i]>=18:
            df['Mayores_Hombres'][i] = 1
        elif df['sexo'][i] == 'Femenino' and df['Edad'][i]>=18:
            df['Mayores_Mujeres'][i] = 1
    return df
Mayores()
```

```python
df['Menores_Hombres'] = 0
df['Menores_Mujeres'] = 0
```

```python
def Menores():
    for i in range(contarFilas):
        if df['sexo'][i] == 'Masculino' and df['Edad'][i]<18:
            df['Menores_Hombres'][i] = 1
        elif df['sexo'][i] == 'Femenino' and df['Edad'][i]<18:
            df['Menores_Mujeres'][i] = 1
    return df
Menores()
```

```python
# Se crea una nueva columna, si es fallecido, coloque 1, de lo contrario, coloca 0

df['Fallecidos'] = 0

def fallecidos():
    for i in range(contarFilas):
        if df['recuperado'][i] == 'Fallecido':
            df['Fallecidos'][i] = 1
    return df
fallecidos()
```

```python
# Se crea una nueva columna, si es fallecido, coloque 1, de lo contrario, coloca 0

df['Recuperados'] = 0

def recuperados():
    for i in range(contarFilas):
        if df['recuperado'][i] == 'Recuperado':
            df['Recuperados'][i] = 1
    return df
recuperados()
```

```python
df['Activo'] = 0

def activo():
    for i in range(contarFilas):
        if df['recuperado'][i] == 'Activo':
            df['Activo'][i] = 1
    return df
activo()
```

```python

# En las casillas con datos vacios, se coloca 0

df = df.fillna(0)
```

```python
# Los datos vacios en la columna recuperado, se colocan como sin clasificar

df['recuperado'].replace(
    to_replace=['N/A'],
    value='Sin clasificar',
    inplace=True
)

sin = df[df['recuperado'] == 'Sin clasificar']
sin
```

```python
df['Pendiente'] = 0

def pendiente():
    for i in range(contarFilas):
        if df['recuperado'][i] == 'Sin clasificar':
            df['Pendiente'][i] = 1
    return df
pendiente()
```

```python
suma_contagios = sum(df['Contagios'])
suma_contagios
```

```python
suma_fallecidos = sum(df['Fallecidos'])
suma_fallecidos
```

```python
suma_recuperados = sum(df['Recuperados'])
suma_recuperados
```

```python
suma_activos = sum(df['Activo'])
suma_activos
```

```python

# Se crear resumen por departamento, diferentes estados y se calculan los porcentajes de participación

Depar = df.groupby(['Lugar']).sum().reset_index()

Depar['% contagios'] = Depar['Contagios']/suma_contagios
Depar['% fallecidos'] = Depar['Fallecidos']/suma_fallecidos
Depar['% recuperados'] = Depar['Recuperados']/suma_recuperados
Depar['% activos'] = Depar['Activo']/suma_activos
Depar
```

```python
# Se coloca la infromación que se necesita del analisis

Departamento = pd.concat([Depar['Lugar'], Depar['Contagios'], Depar['% contagios'],  Depar['Fallecidos'], Depar['% fallecidos'], Depar['Recuperados'], Depar['% recuperados'], Depar['Activo'], Depar['% activos'], Depar['Pendiente']], axis=1)

Departamento.columns = ['Departamento', 'Contagios', '% contagiados','Fallecidos', '% fallecidos', 'Recuperados', '% recuperados', 'Activos', '% activos', 'Pendientes clasificar']

Departamento 
```

```python
# Se crear resumen por departamento y municipios con diferentes estados

Mun = df.groupby(['Lugar', 'ciudad_municipio_nom']).sum().reset_index()

```

```python
# Se crear resumen por departamento, diferentes estados y se calculan los porcentajes de participación

Mun['% contagios'] = Mun['Contagios']/suma_contagios
Mun['% fallecidos'] = Mun['Fallecidos']/suma_fallecidos
Mun['% recuperados'] = Mun['Recuperados']/suma_recuperados
Mun['% activos'] = Mun['Activo']/suma_activos
Mun
```

```python
Municipio = pd.concat([Mun['Lugar'], Mun['ciudad_municipio_nom'], Mun['Contagios'], Mun['% contagios'], Mun['Fallecidos'], Mun['% fallecidos'], Mun['Recuperados'], Mun['% fallecidos'], Mun['Activo'], Mun['% activos'],Mun['Pendiente']], axis=1)

Municipio.columns = ['Departamento', 'Municipio', 'Contagios', '% contagios', 'Fallecidos', '% fallecidos', 'Recuperados', '% fallecidos', 'Activos', '% activos', 'Pendientes clasificar']

Municipio 
```
